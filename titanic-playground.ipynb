```python
# Verifique se há valores ausentes
df.isnull().sum()
```

```python
# Remova as linhas com valores ausentes
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Troque os valores binários por string para indicar se a pessoa sobreviveu ou não
df['2urvived'] = df['2urvived'].apply(lambda x: 'Yes' if x == 1 else 'No')
```

```python
df['2urvived']
```

```python
# Adicione uma coluna para indicar a idade em dias
df['Age_days'] = df['Age'] * 365
```

```python
df['Age_days'] 
```

```python
# Agrupe os dados por gênero e calcule a média da idade
grouped = df.groupby(['Sex'])['Age'].mean()

grouped
```

```python
# Selecione apenas as linhas onde a idade é maior que 30 anos
df = df[df['Age'] > 30]

df
```

```python
import matplotlib.pyplot as plt

# Crie um gráfico de barras para mostrar a distribuição de gênero
df['Sex'].value_counts().plot(kind='bar')

plt.show()
```
