# Usando Matplotlib

```python
# importando o pacote
import matplotlib.pyplot as plt
```

## 1. Gr√°fico de Linha

```python
# Dados para as coordenadas do gr√°fico
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gr√°fico de linha')

plt.show()
```

## 2. Gr√°fico de Barras

```python
# Dados para as barras
categorias = ['Categoria 1', 'Categoria 2', 'Categoria 3']
valores = [10, 20, 30]

plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Gr√°fico de barra')

plt.show()
```

## 3. Gr√°fico de Pizza (ou Torta)

```python
# Dados para as fatias
categorias = ['Categoria 1', 'Categoria 2', 'Categoria 3']
valores = [10, 20, 30]

plt.pie(valores, labels=categorias)
plt.title('Gr√°fico de pizza')

plt.show()
```

## 4. Gr√°fico de Dispers√£o

```python
# Dados para as coordenadas
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gr√°fico de dispers√£o')

plt.grid()
```
